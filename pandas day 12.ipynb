```python
# to generate a dommy data: any random data
# create a synthetic student score data / generate a table of ca score for 50 student 

column_name = ['ca1', 'ca2', 'ca3']
sit_no = np.arange (1,51)
np.random.seed(2)
score = np.random.randint(1,11,150)
score = score.reshape(50,3)
student_score = pd.DataFrame(data = score, index = sit_no, columns = column_name)
student_score 
```

```python
# to insert a new column : this will add at the end
# add a column called exam score
np.random.seed(2)
student_score['exam score'] = np.random.randint(25,71,50).reshape(50,1)
student_score
```

```python
# to insert a new column at choice location
# add a column called age: oldest 16, youngest 13
student_score.insert(loc = 0, value = np.random.randint(13,17,50).reshape(50,1), column = 'age')
student_score
```

```python
# create a new column from existing column
# add a column called total score: that is ca1+ca2+ca3+exan score
student_score['total score'] = student_score['ca1'] + student_score['ca2'] + student_score['ca3'] + student_score['exam score']
student_score
```

```python
# create a new column from existing column by applying a function
# create a column called grade 
def grade(totalscore):
    if totalscore >= 70 and totalscore <=  100:
        return 'A'
    elif totalscore >= 60 and totalscore <=  69.99:
        return 'B'
    elif totalscore >= 50 and totalscore <=  59.99:
        return 'C'
    elif totalscore >= 40 and totalscore <=  49.99:
        return 'D'
    elif totalscore >= 30 and totalscore <=  39.99:
        return 'E'
    else:
        return 'F'
student_score['grade'] = student_score['total score'].apply(grade)
student_score
```

```python
# before we use any given data, we must first check extention: 
# to check the extention of a data
# df means data frame
df = pd.read_csv('sample_pivot.csv')
df.head
```

```python
# check the information;
df.info()
# glancing through the info, you can notice that unit which is suppose to be 1000 is not complete
```

```python
# to check for missing values
df.isnull().sum()
# here we have 89 missing values
```

```python
# to tackle missing values
# we must fill them by diff technique
# the first two technique should be used only when data are very related else use others like statistical techniques
df.tail()
```

```python
# 1, forward fill technique: will replicate the data behind the empty column in the empty column
df1 = df.copy()
df1.fillna(method = 'ffill', inplace = True)
df1.tail()
```

```python
# 2, backward fill
df1 = df.copy()
df1.fillna(method = 'bfill', inplace = True)
df1.tail()
```

```python
# 3, statistical 
df2 = df.copy()
df2['Units'].fillna(df2['Units'].mean(),inplace = True)
df2.tail()
```

```python
# DROPPING OF MISSING VALUES: this will drop the entire roll where a value is missing
df.dropna(inplace = True)
df.isnull().sum()
```
